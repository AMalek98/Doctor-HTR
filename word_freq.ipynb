```python
import pandas as pd
from collections import Counter

# Path to the text file
file_path = "dwe_list.txt"

# Read the file and process the lines
with open(file_path, 'r') as file:
    lines = file.readlines()

# Extract the words after the colon
words = []
for line in lines:
    if ':' in line:
        medicine_name = line.split(':', 1)[1].strip()
        words.extend(medicine_name.split())

# Count the frequency of each word
word_counts = Counter(words)

# Create the DataFrame
df = pd.DataFrame(word_counts.items(), columns=['medicine', 'freq'])

# Display the DataFrame
print(df)
```

```python
import matplotlib.pyplot as plt

# Get the top 5 most frequent medicines
top_5 = df.nlargest(5, 'freq')

# Plotting
plt.figure(figsize=(10, 6))
bars = plt.bar(top_5['medicine'], top_5['freq'], color='skyblue')

# Adding the frequency values on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom')

plt.xlabel('Medicine')
plt.ylabel('Frequency')
plt.title('Top 5 Most Frequent Medicines')
plt.show()
```

```python
import matplotlib.pyplot as plt
# Get the top 5 most frequent medicines
bot_5 = df.nsmallest(5, 'freq')

# Plotting
plt.figure(figsize=(10, 6))
bars = plt.bar(bot_5['medicine'], bot_5['freq'], color='orange')

# Adding the frequency values on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, int(yval), ha='center', va='bottom')

plt.xlabel('Medicine')
plt.ylabel('Frequency')
plt.title('Top 5 Most Frequent Medicines')
plt.show()
```
